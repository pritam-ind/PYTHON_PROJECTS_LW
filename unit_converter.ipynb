```python
# print("Welcome to Unit Converter")
print("..........................")
print(" Enter 1 for Temperature ")
print(" Enter 2 for Currency ")
print(" Enter 3 for Length ")
print(" Enter 4 for Weight ")
print("..........................\n")

choice = input("Please Enter the correct choice to proceed : ")
    
if choice == "1":
    print("Temperature Conversion")
    print("       ......       ")
    print("Enter 1 for Celsius to Fahrenheit")
    print("Enter 2 for Fahrenheit to Celcius")
    
    op = input("\nSelect an to proceed option: ")
    
    if op=="1":
        c = float(input("Enter temp in Celsius: ")) 
        f = (c * 1.8)+ 32 
        print(f"Temperature in Fahrenheitt is: {f}" )
    elif op=="2":
        f = float(input("Enter temp in Fahrenheit: "))
        c = (f - 32) /1.8
        print(f"Temperature in Celsius is: {c}")
    else :
        print("You Selected Invalid option")
        
elif choice=="2":
    print("       ......       ")
    print("Currency Conversion")
    print("       ......       \n")
    print("Press 1 for Rupee to Dollar")
    print("Press 2 for Dollar to Rupee")
    
    op = input( "\nPlease Select the correct option to proceed : ")
    if op == "1":
        rupee = float(input(" Enter amount in Rupees: "))
        dollar = rupee/86.5
        print(f"Amount in Dollar is:{dollar}")
    elif op =="2":
        dollar = float(input("Enter amount in Dollars: "))
        rupee = dollar *86.5
        print(f"Amount in Rupees is:{rupee}" )
    else :
        print("You select the Invalid option! ")

elif choice=="3":
    print("       ......       ")
    print("Length Conversion")
    print("       ......       \n")
    print("Press 1 for Km to  Mile ")
    print("Press 2 for Mile to  Km")
    
    op = input("Please Select the option: ")
    if op== "1":
        km = float(input("Enter the value in  Km: "))
        mile = km*0.621371
        print(f"The Value in Miles is: {mile}")
    elif op =="2":
        mile = float(input("Enter value in Miles: "))
        km = mile / 0.621371
        print(f"Value in Km is: {km}")
    else :
        print("You selected Invalid option")
        
elif choice =="4":
    print("       ......       ")
    print("Weight Conversion")
    print("       ......       \n")
    print("Press 1 for Kg to Pound")
    print("Press 2 for Pound to Kg")
     
    op = input("Select option: ")
    
    if op =="1": 
        kg = float(input("Enter weight in Kg: "))
        pound = kg * 2.20462
        print("Weight in Pounds is:", pound)
    elif op == "2":
        pound = float(input("Enter weight in Pounds: "))
        kg = pound / 2.20462
        print("Weight in Kg is:", kg) 
    else :
        print("You Entered a Invalid option") 
        
else :
    print("Opppps....... Wrong choice")
```
