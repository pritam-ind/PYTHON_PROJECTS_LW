```python
# Project: Email Sender using SMTP
import smtplib
import ssl
from email.message import EmailMessage

SENDER = "mysxxxxxxxxxm1@gmail.com"
PASSWORD = "XXXXXXXXXXXXXXXXXXXXX" # enter your app password here
RECEIVER = "souvxxxxxxxx25@gmail.com"

msg = EmailMessage()
msg["From"] = SENDER
msg["To"] = RECEIVER
msg["Subject"] = "Automated Email from Python"
msg.set_content("Hello, I am Pritam. This email was sent using Python script.")

context = ssl.create_default_context()

try:
    with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as server:
        server.login(SENDER, PASSWORD)
        server.send_message(msg)
    print("Email Sent Successfully!")

except Exception as e:
    print("Error:", e)
```
