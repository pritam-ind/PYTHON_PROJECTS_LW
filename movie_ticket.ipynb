```python
#movie ticket 
# import time
age = int(input("Enter your age: "))
if age > 18:
    print("\nYou are eligible to watch movie.")
    ticket = input("Do you have ticket?\nSelect Yes/No: ").lower().strip()
    if ticket == "yes":        
        print("Please go and enjoy the movie")
    else:
        print("Please go and book a ticket first to enjoy the movie!")

elif 14 < age < 18:
    print("You are not allowed but can watch movie under the guidence of your parents!")
    withParent = input("\nAre you with your parents? \nSelect yes/no: ")
    if withParent == "yes":
        print("Go with your parent(s) and enjoy the movie!")
    else:
        print("Please go and book a ticket first to enjoy the movie with your parents!")
        
else:
    print("Sorry sir, you are not allowed!")
```
