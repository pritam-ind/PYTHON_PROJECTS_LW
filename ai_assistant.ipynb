```python
import speech_recognition as sr
import pyttsx3
import webbrowser
import os
# import time

speaker = pyttsx3.init()
mic = sr.Recognizer()
def speak(text):
    """Helper function to make speaking easier"""
    print(f"Jarvis: {text}")
    speaker.say(text)
    speaker.runAndWait()

def take_command():
    """Listens to the microphone and returns text"""
    try:
        with sr.Microphone() as source:
            print("\nListening...")
            mic.adjust_for_ambient_noise(source) 
            voice = mic.listen(source)
            command = mic.recognize_google(voice)
            return command.lower() 
    except:
        return "none"

if __name__ == "__main__":
    
    speak("Hello Pritam Sir. I am online. What can I do for you?")

    while True:
        text = take_command()
        
        if text == "none":
            continue 

        print(f"User said: {text}")
        
        if "open chrome" in text or "open google" in text:
            speak("Opening Google Chrome...")
            
            webbrowser.open("www.google.com")

        elif "open netflix" in text:
            speak("Opening Netflix for you, Sir.")
            webbrowser.open("www.netflix.com")

        elif "open gemini" in text:
            speak("Opening Google Gemini AI...")
            webbrowser.open("gemini.google.com")

        elif "bluetooth" in text:
            speak("Opening Bluetooth Settings...")
            os.system("start ms-settings:bluetooth")

        elif "notepad" in text:
            speak("Opening Notepad...")
            os.system("notepad")

        elif "exit" in text or "stop" in text or "bye" in text:
            speak("Goodbye Pritm Sir. Have a nice day, See you again!")
            break
```
