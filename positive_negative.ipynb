```python
while True:
    print("")
    enter_value = input("Enter a integer to check whether a number is positive or negative:\nEnter exit to stop: \n ")
    if enter_value.lower() == "exit":
        break
    elif int(enter_value) == 0:
        print(f"You entered {enter_value}, which is neither positive nor negative")
    elif int(enter_value) < 0:
        print(f"You entered {enter_value} which is Negative")
    else:
        print(f"You entered {enter_value} which is Positive")
```
